```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
# Box plots helps us finding percentiles

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# selectting a normal distibution b/w a lower value(1 in this case)
# and the std deviation with a step size of 100

print(data)

```

```python
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()

#circles are basically outliers
```

```python
plt.boxplot(data,vert=False,patch_artist=False)
plt.show()
```

# Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210] 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True) #indicates the format we want the value to ne shown
# we have chosen to be a float of 1 decimal.

plt.axis('equal')
plt.show()
```
